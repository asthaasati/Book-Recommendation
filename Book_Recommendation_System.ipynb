Popularity based Recommendation

```python
Ratings_with_name=Ratings.merge(Books,on='ISBN')
```

```python
num_rating_df=Ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
print(Ratings_with_name['Book-Rating'].unique())
```

```python
Ratings_with_name['Book-Rating'] = pd.to_numeric(Ratings_with_name['Book-Rating'], errors='coerce')
```

```python
Ratings_with_name = Ratings_with_name.dropna(subset=['Book-Rating'])
```

```python
Ratings_with_name['Book-Rating'] = Ratings_with_name['Book-Rating'].fillna(0)
```

```python
avg_rating_df = Ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating': 'avg_ratings'}, inplace=True)
```

```python
avg_rating_df
```

```python
popular_df=num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings',ascending=False).head(101)
```

```python
popular_df.merge(Books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
```

Collaborative filtering based Recommendation

```python
x=Ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
x[x]
```

```python
Readers=x[x].index
```

```python
filtered_rating=Ratings_with_name[Ratings_with_name['User-ID'].isin(Readers)]
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
```

```python
famous_books=y[y].index
```

```python
famous_books
```

```python
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
import pandas as pd
import pickle

# Load your raw dataset
raw_data = pd.read_csv('Books.csv', low_memory=False)

# Check if required columns exist
if all(col in raw_data.columns for col in ['num_ratings', 'avg_rating']):
    print("Required columns already exist. Proceeding...")
elif 'Ratings' in raw_data.columns:
    print("Generating missing columns from 'Ratings'...")
    ratings_grouped = raw_data.groupby('Book-Title')['Ratings']
    raw_data['num_ratings'] = ratings_grouped.transform('count')
    raw_data['avg_rating'] = ratings_grouped.transform('mean')
else:
    print("Required columns or data to generate them are missing. Skipping these columns.")
    raw_data['num_ratings'] = None
    raw_data['avg_rating'] = None

# Preprocess the data to create the required columns
available_columns = ['Book-Title', 'Book-Author', 'Image-URL-M', 'num_ratings', 'avg_rating']
popular_df = raw_data[[col for col in available_columns if col in raw_data.columns]].copy()

# Handle missing values
if 'Image-URL-M' in popular_df.columns:
    popular_df['Image-URL-M'] = popular_df['Image-URL-M'].fillna('default.jpg')
popular_df = popular_df.dropna(subset=['Book-Title', 'Book-Author'])

# Save the processed data
with open('popular.pkl', 'wb') as file:
    pickle.dump(popular_df, file)

print("Processed popular.pkl successfully created!")
```
